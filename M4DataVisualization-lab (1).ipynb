## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# your code goes here
QUERY = """
        SELECT ConvertedComp FROM master
        """
df = pd.read_sql(QUERY, conn)
sns.histplot(data=df, x="ConvertedComp", kde=True)
plt.title("Distribution of Converted Compensation")
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
        SELECT Age FROM master
        """
        
df = pd.read_sql(QUERY, conn)
plt.figure(figsize=(4, 8))
sns.boxplot(data=df["Age"])
plt.title("Box Plot of Age")
plt.ylabel("Age")
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
        SELECT Age, WorkWeekHrs FROM master
        """

df = pd.read_sql(QUERY, conn)
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="WorkWeekHrs", y="Age")
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
# your code goes here
QUERY = """
        SELECT Age, WorkWeekHrs FROM master
        """

df = pd.read_sql(QUERY, conn)
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="WorkWeekHrs", y="Age", size="Age")
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
        SELECT DatabaseDesireNextYear, COUNT(DatabaseDesireNextYear) AS OCCURENCES FROM DatabaseDesireNextYear GROUP BY DatabaseDesireNextYear ORDER BY OCCURENCES DESC
        """
df = pd.read_sql(QUERY, conn)
df = df.head()
plt.pie(x=df["OCCURENCES"], labels=df["DatabaseDesireNextYear"], autopct='%1.1f%%')
plt.title("Share of different databases desired by developers for next year")
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """SELECT WorkWeekHrs, CodeRevHrs FROM master WHERE Age BETWEEN 30 AND 35 GROUP BY Age"""
plt.figure(figsize=(12, 6))
df = pd.read_sql(QUERY, conn)
df.plot(kind='bar', stacked=True)
age = [30, 31, 32, 33, 34, 35]
plt.title("Stacked Chart")
plt.xticks(ticks=[0, 1, 2, 3, 4, 5], labels=age)
plt.xlabel("Age")
plt.ylabel("Hours")
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """SELECT Age, ConvertedComp FROM master WHERE Age BETWEEN 45 AND 60 GROUP BY Age"""

df = pd.read_sql(QUERY, conn)
sns.lineplot(data=df, x=df["Age"], y=df["ConvertedComp"])
plt.title("Line Plot of Median Compensation Per Age")
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """SELECT COUNT(Respondent) as count, MainBranch FROM master GROUP BY MainBranch"""

df = pd.read_sql(QUERY, conn)
sns.barplot(data=df, x=df["count"], y="MainBranch")
```
